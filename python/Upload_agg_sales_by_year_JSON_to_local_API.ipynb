```python
from flask import Flask, jsonify
import json
app = Flask(__name__)
with open('./data_to_upload_through_Azure_Data_Factory/agg_sales_by_year.json',"r",encoding="utf-8") as f:
    data = json.load(f)

@app.route("/agg_sales_by_year", methods=["GET"])
def products():
    return jsonify(data)

@app.route("/health", methods=["GET"])
def health():
    return {"status": "ok", "count": len(data)}

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000)


###  http://localhost:5000/health → should show {"status":"ok","count":21}
###  http://localhost:5000/agg_sales_by_year → JSON complete
```
