```python
# server.py
from flask import Flask, jsonify
import json
app = Flask(__name__)
with open("products_109.json","r",encoding="utf-8") as f:
    data = json.load(f)

@app.route("/products", methods=["GET"])
def products():
    return jsonify(data)

@app.route("/health", methods=["GET"])
def health():
    return {"status": "ok", "count": len(data)}

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000)


###  http://localhost:5000/health → debería devolver {"status":"ok","count":109}
###  http://localhost:5000/products → JSON completo
```

```python
import requests
import pandas as pd

API_URL = "http://localhost:5000/agg_sales_by_year"   # endpoint local (Flask)

# ---------- GET from local API ----------
resp = requests.get(API_URL, timeout=20)
print("resp:\s",resp)
resp.raise_for_status()
agg = resp.json()
print("agg:\s", agg)
df = pd.DataFrame(agg)
print("Read from local API:", len(df))
print(df.head())
```
