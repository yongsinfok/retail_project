```python
# extract_products_to_json
import pandas as pd
import pyodbc
import json

# Adjust conection
conn_str = "DRIVER={ODBC Driver 17 for SQL Server};SERVER=localhost\\SQLEXPRESS;DATABASE=retail_project;Trusted_Connection=yes;"
cnxn = pyodbc.connect(conn_str)
sql = """
SELECT DISTINCT PRODUCTCODE AS product_id,
       PRODUCTLINE AS product_category,
       MSRP AS product_price
FROM dbo.sales_data_raw;
"""

df = pd.read_sql(sql, cnxn)
cnxn.close()

df['id'] = df.index + 1  # generate ids 1..N

print("dataframe with products data", df.head())

data = df.to_dict(orient='records')
with open("products_109.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("JSON guardado en products_109.json, filas:", len(data))
```
